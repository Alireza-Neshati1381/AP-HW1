```python
from abc import ABC, abstractmethod

class Vehicles:
    def __init__(self, kind, brand, model, fuel, path):
        self._kind = kind
        self.brand = brand
        self.model = model
        self.fuel = fuel
        self._path = path
    def get_kind(self):
        return self._kind
    
    def get_path(self):
        return self._path
    
    def move(self) :
        return "The vehicle is moving."
        
    def start_engine(self):
        return "The vehicle started."
        
    def stop_engine(self):
        return "The vehicle stoped."

class public_vehicles(Vehicles,ABC):
    def __init__(self, kind, brand, model, fuel, path, ticket_price, standing, electronic_card):
        super().__init__(kind, brand, model, fuel, path)
        self.ticket_price = ticket_price
        self.standing = standing
        self.electronic_card = electronic_card
    
    def standing_status(self):
        if self.standing == True :
            return "In this vehicle, you can stand."
        else :           
            return "In this vehicle, you can't stand."
    
    def public(self):
        return "All these vehicles are public."
    
    def sitting(self):
        return "In the vehicle, you can sit."
    
    @abstractmethod
    def station(self):
        return "The vehicle has station over the city or no!"
    
    @abstractmethod
    def ticket_reserve_status(self):
        return "You should reserve ticket for this vehicle."
    
class private_vehicles(Vehicles,ABC):
    def __init__(self, kind, brand, model, fuel, path, price, start, number_of_seats):
        super().__init__(kind, brand, model, fuel, path)
        self.price = price
        self.start = start
        self.number_of_seats = number_of_seats
    
    @abstractmethod    
    def introduction(self):
        return f"This is a/an {self.brand} {self.model}"
    
    @abstractmethod
    def parking_place(self):
        return "You can park this vehicle where that is ligal."
    
    @abstractmethod
    def get_fuel(self):
        return self.fuel
    
    def private(self):
        return "All these vehicles are private."
    
    def start(self) :
        return "The vehicle is started"
        
    def lock(self) :
        return "The vehicle is locked"
    
    def unlock(self) :
        return "The vehicle is unlockeed"
   
class car(private_vehicles):
    def __init__(self, kind, brand, model, fuel, path, price, start, number_of_seats, Type, zero_to_hundred, gear):
        super().__init__(kind, brand, model, fuel, path, price, start, number_of_seats)
        self.Type = Type
        self.zero_to_hundred = zero_to_hundred
        self.gear = gear
    
    def introduction(self):  # overriding the start_status method.
        return f"This is a/an {self.brand} {self.model}"
    
    def parking_place(self):   # overriding the parking_place method.
        return "You can park your car anywhere that is ligal."
    
    def get_fuel(self):    # overriding the parking_place method.
        return self.fuel
    
    def start_status(self):   
        if self.start == "key" :
            return f"a/an {self.brand} {self.model} use's key to start."
        elif self.start == "buttom" :
            return f"a/an {self.brand} {self.model} use's buttom to start."
    
    def gear_type(self):
        if self.gear == "automatic" :
            return f"a/an {self.brand} {self.model} is an automatic car."
        elif self.gear == "manual" :
            return f"a/an {self.brand} {self.model} is a manual car"

class motorcycle(private_vehicles):
    def __init__(self, kind, brand, model, fuel, path, price, start, number_of_seats, zero_to_hundred, gear):
        super().__init__(kind, brand, model, fuel, path, price, start, number_of_seats)
        self.zero_to_hundred = zero_to_hundred
        self.gear = gear
    
    def introduction(self):    # overriding the parking_place method.
        return f"This is a/an {self.brand} {self.model}"
    
    def parking_place(self):   # overriding the parking_place method.
        return "You can park your motorcycle anywhere that is ligal easily."

    def get_fuel(self):    # overriding the parking_place method.
        return self.fuel
    
    def gear_type(self):
        if self.gear == "automatic" :
            return f"a/an {self.brand} {self.model} is an automatic motorcycle."
        elif self.gear == "manual" :
            return f"a/an {self.brand} {self.model} is a manual motorcycle"

class private_jet(private_vehicles):
    def __init__(self, kind, brand, model, fuel, path, price, start, number_of_seats, engine, length):
        super().__init__(kind, brand, model, fuel, path, price, start, number_of_seats)
        self.engine = engine
        self.length = length
    def introduction(self):    # overriding the parking_place method.
         return f"This is a/an {self.brand} {self.model}"
     
    def parking_place(self):   # overriding the parking_place method.
         return "You can park your motorcycle anywhere that is ligal easily."

    def get_fuel(self):    # overriding the parking_place method.
         return self.fuel
     
    def rider(self):
        return "Private jet needs pilot"
    
    
    
    
```
