```python
from abc import ABC, abstractmethod

class Vehicles:
    def __init__(self, kind, brand, model, fuel, path):
        self._kind = kind  # Encapsulation
        self.brand = brand
        self.model = model
        self.fuel = fuel
        self._path = path  # Encapsulation
        
    def get_kind(self):   # Encapsulation
        return self._kind
    
    def get_path(self):   # Encapsulation
        return self._path
    
    def move(self) :
        return "The vehicle is moving."
        
    def start_engine(self):
        return "The vehicle started."
        
    def stop_engine(self):
        return "The vehicle stoped."


class Public_vehicles(Vehicles,ABC):
    def __init__(self, kind, brand, model, fuel, path, ticket_price, standing = None, electronic_card = None):
        super().__init__(kind, brand, model, fuel, path)
        self.ticket_price = ticket_price
        self.standing = standing
        self.electronic_card = electronic_card
    
    def standing_status(self):
        if self.standing == True :
            return "In this vehicle, you can stand."
        else :           
            return "In this vehicle, you can't stand."
    
    def public(self):
        return "All these vehicles are public."
    
    def sitting(self):
        return "In this vehicle, you can sit."
    
    @abstractmethod     # Abstraction
    def station(self):
        return "The vehicle has station over the city or no!"
    
    @abstractmethod     # Abstraction
    def ticket_reserve_status(self):
        return "You should reserve ticket for this vehicle."
    
    
class Private_vehicles(Vehicles,ABC):
    def __init__(self, kind, brand, model, fuel, path, price, start, number_of_seats):
        super().__init__(kind, brand, model, fuel, path)
        self.price = price
        self.start = start
        self.number_of_seats = number_of_seats
    
    @abstractmethod    # Abstraction
    def introduction(self):
        return f"This is a/an {self.brand} {self.model}"
    
    @abstractmethod    # Abstraction
    def parking_place(self):
        return "You can park this vehicle where that is ligal."
    
    @abstractmethod    # Abstraction
    def get_fuel(self):
        return self.fuel
    
    def private(self):
        return "All these vehicles are private."
    
    def start(self) :
        return "The vehicle is started"
        
    def lock(self) :
        return "The vehicle is locked"
    
    def unlock(self) :
        return "The vehicle is unlockeed"
    
    
class Engine:   
    def __init__(self, code_number, capacity, number_of_cylinders, horsepower, piston):
        self._code_Number = code_number
        self.capacity = capacity
        self.number_of_cylinders = number_of_cylinders
        self.horsepower = horsepower
        self.piston = piston
         
        
    def get_code_number(self): # Encapsulation 
        return self._Code_Number
    
    def Horsepower(self):
        return f"{self.model} {self.brand} has f{self.horsepower} horsepower."
    
    def Type_by_cylinder(self):
        return f"This is a V{self.Number_of_cylinders} Car !"
    
    def Capacity(self):
        return self.capacity
    
    
class Car(Private_vehicles):
    def __init__(self, kind, brand, model, fuel, path, price, start, number_of_seats, zero_to_hundred, gear, 
                 Code_Number, Capacity, Number_of_cylinders, Horsepower, piston):    
        super().__init__(kind, brand, model, fuel, path, price, start, number_of_seats)
        self.zero_to_hundred = zero_to_hundred
        self.gear = gear
        self.obj_engine = Engine(Code_Number, Capacity, Number_of_cylinders, Horsepower, piston)
    
    def introduction(self):  # overriding the start_status method.
        return f"This is a/an {self.model} {self.brand}."
    
    def parking_place(self):   # overriding the parking_place method.
        return "You can park your car anywhere that is ligal."
    
    def get_fuel(self):    # overriding the parking_place method.
        return self.fuel
    
    def start_status(self):   
        if self.start == "key" :
            return f"a/an {self.model} {self.brand} use's key to start."
        elif self.start == "buttom" :
            return f"a/an {self.model} {self.brand} use's buttom to start."
    
    def gear_type(self):
        if self.gear == "automatic" :
            return f"a/an {self.model} {self.brand} is an automatic car."
        elif self.gear == "manual" :
            return f"a/an {self.model} {self.brand} is a manual car"
    
    def engine_code(self):  # composition
        return self.obj_engine.get_code_number()
    
    def car_horsepower(self):   # composition
        return self.obj_engine.Horsepower()


class Motorcycle(Private_vehicles):
    def __init__(self, kind, brand, model, fuel, path, price, start, number_of_seats, zero_to_hundred, gear):
        super().__init__(kind, brand, model, fuel, path, price, start, number_of_seats)
        self.zero_to_hundred = zero_to_hundred
        self.gear = gear
    
    def introduction(self):    # overriding the parking_place method.
        return f"This is a/an {self.brand} {self.model}"
    
    def parking_place(self):   # overriding the parking_place method.
        return "You can park your motorcycle anywhere that is ligal easily."

    def get_fuel(self):    # overriding the parking_place method.
        return self.fuel
    
    def gear_type(self):
        if self.gear == "automatic" :
            return f"a/an {self.model} {self.brand} is an automatic motorcycle."
        elif self.gear == "manual" :
            return f"a/an {self.model} {self.brand} is a manual motorcycle"


class Private_jet(Private_vehicles):
    def __init__(self, kind, brand, model, fuel, path, price, number_of_seats, engine, length):
        super().__init__(kind, brand, model, fuel, path, price, number_of_seats)
        self.engine = engine
        self.length = length
        
    def introduction(self):    # overriding the parking_place method.
         return f"This is a/an {self.brand} {self.model}"
     
    def parking_place(self):   # overriding the parking_place method.
         return "You should park your jet in airport."

    def get_fuel(self):    # overriding the parking_place method.
         return self.fuel
     
    def rider(self):
        return "Private jet needs pilot"
    
    def luxury(self):
        return "Private jet is a luxury vehicle."
    

class Helicopter(Private_vehicles):
    def __iniit__(self, kind, brand, model, fuel, path, price, number_of_seats, pilot):
        super().__init__(kind, brand, model, fuel, path, price, number_of_seats)
        self.pilot = pilot
        
    def introduction(self):    # overriding the parking_place method.
        return f"This is a/an {self.brand} {self.model}"
     
    def parking_place(self):   # overriding the parking_place method.
        return "You should park your helicopter on a helipad."

    def get_fuel(self):    # overriding the parking_place method.
        return self.fuel
     
    def luxury(self):
        return "Helicopter is a luxury vehicle."
    
    
class Ticket_reservation:   # Aggregation
    def __init__(self, name, phone_number, origin, destination, vehicle_type, p_airplane_ticket, p_ship_ticket, 
                 p_train_ticket, time_price):
        self.name = name
        self.phone_number = phone_number
        self.origin = origin
        self.destination = destination
        self._vehicle_type = vehicle_type   # Encapsulation
        self.p_ship_ticket = p_ship_ticket
        self.p_airplane_ticket = p_airplane_ticket
        self.p_train_ticket = p_train_ticket
        self.time_price = time_price
    
    def customer_information(self):
        return f"name : {self.name}\nphone_number : {self.phone_number}"
    
    def vehicle(self):
        return self._vehicle_type()
    
    def origin_destination(self):
        return f"Origin : {self.origin}\nDestination : {self.destination}"
    
    def best_choice(self):   
        if self.time_price == "time" :     
            return "Airplane is the best choice."
        elif self.time_price == "price" :
            if self.p_train_ticket > self.p_ship_ticket :
                return "Train is the best choice."
            else:
                return "Ship is the best choice."
    

class Airplane(Public_vehicles):
    def __init__(self, kind, brand, model, fuel, path, ticket_price, standing, engine, length, Type, ticket_reservation):
        super().__init__(kind, brand, model, fuel, path, ticket_price, standing)
        self.engine = engine
        self.length = length
        self.Type = Type
        self.obj_ticket_reservation = ticket_reservation # Aggregation
        
    def station(self):      # overriding the station method.
        return "Airplane needs airport!"
          
    def ticket_reserve_status(self):     # overriding the ticket reserve status.
        return "You should reserve ticket for airplane."
    
    def runway(self):
        return "Airplane needs runway."
    
    def taking_off(self):
        return "You should fasten your seat belts in this situation."
    
    def landing(self):
        return "You should fasten your seat belts in this situation."
    
    def get_Type(self):
        if self.Type == "freight" :
            return "Freight airplane!"
        elif self.Type == "passenger" :
            return "Passenger airplane!"
        
    def recommendation(self):   # Aggregation
        return self.obj_ticket_reservation.best_choice()
    
    
class Bus(Public_vehicles):
    def __init__(self, kind, brand, model, fuel, path, standing, electronic_card, number_of_seats, Type):
        super().__init__(kind, brand, model, fuel, path, standing, electronic_card)
        self.number_of_seats = number_of_seats
        self.Type = Type
        
    def station(self):      # overriding the station method.
        return "Bus has station over the city!"
          
    def ticket_reserve_status(self):     # overriding the ticket reserve status.
        if self.electronic_card == False :
            return "You should reserve ticket for airplane."
        
    def pay(self):
        return "you can pay by cash or electronic card."
    
    def bus_type(self):
        if self.Type == "inter-city" :
            return "You can use this kind of bus for traveling."
        elif self.Type == "city" :
            return "This type of bus is city bus."
    
    def get_Type(self):
        return self.Type
    
    
class Subway(Public_vehicles):
    def __init__(self, kind, brand, model, fuel, path, ticket_price, standing, electronic_card, 
                 emergency_break_key, hammer, station_name):
        super().__init__(kind, brand, model, fuel, path, ticket_price, standing, electronic_card)
        self.emergency_break_key = emergency_break_key
        self.hammer = hammer
        self._station_name = station_name    # Encapsulation
        
    def station(self):      # overriding the station method.
        return "Subway has station over the city!"
          
    def ticket_reserve_status(self):     # overriding the ticket reserve status.
        if self.electronic_card == False :
            return "You should reserve ticket for subway."
     
    def pay(self):
        if self.electronic_card == True :
            return "You should pay by electronic card."
        
    def emergency_break(self):
        if self.emergency_break_key == True:
            return "You can use this key for emergency situation to stop the subway."
    
    def emergency_hammer(self):
        if self.hammer == True:
            return "You can use hammer for emergency situation to break windows."
        
    def operator_sound(self):   # Encapsulation
        return f"{self._station_name} station."


class Train(Public_vehicles):
    def __init__(self, kind, brand, model, fuel, path, ticket_price, standing, electronic_card, Type, ticket_reservation):
        super().__init__(kind, brand, model, fuel, path, ticket_price, standing, electronic_card)
        self.Type = Type
        self.obj_ticket_reservation = ticket_reservation # Aggregation
        
    def station(self):      # overriding the station method.
        return "Train has station over the city!"
          
    def ticket_reserve_status(self):     # overriding the ticket reserve status.
        if self.electronic_card == False :
            return "You should reserve ticket for train."
     
    def get_ticket_price(self):
        return self.ticket_price
    
    def sound_of_train(self):
        return "hoo hoo chi chi!"
    
    def get_Type(self):
        if self.Type == "freight" :
            return "Freight train!"
        elif self.Type == "passenger" :
            return "Passenger train!"
        
    def recommendation(self):   # Aggregation
        return self.obj_ticket_reservation.best_choice()
    
    
class Ship(Public_vehicles):
    def __init__(self, kind, brand, model, fuel, path, ticket_price, standing, electronic_card, Type, ticket_reservation):
        super().__init__(kind, brand, model, fuel, path, ticket_price, standing, electronic_card)
        self.Type = Type
        self.obj_ticket_reservation = ticket_reservation # Aggregation
        
    def station(self):      # overriding the station method.
        return "Ship needs wharf!"
          
    def ticket_reserve_status(self):     # overriding the ticket reserve status.
        if self.electronic_card == False :
            return "You should reserve ticket for ship."
    
    def get_Type(self):
        if self.Type == "freight" :
            return "Freight ship!"
        elif self.Type == "passenger" :
            return "Passenger ship!"
    
    def get_ticket_price(self):
        return self.ticket_price
    
    def recommendation(self):   # Aggregation
        return self.obj_ticket_reservation.best_choice()
    
    
class Taxi(Public_vehicles):
    def __init__(self, kind, brand, model, fuel, path, standing, electronic_card, number_of_seats, random_fee_per_km, random_distance_km):
        super().__init__(kind, brand, model, fuel, path, standing)
        self.number_of_seats = number_of_seats
        self.random_distance_km = random_distance_km
        self.__random_fee_per_km = random_fee_per_km  # Encapsulation
        
    def station(self):      # overriding the station method.
        return "Taxi has station over the city!"
          
    def ticket_reserve_status(self):     # overriding the ticket reserve status.
        if self.electronic_card == False :
            return "You should reserve ticket for taxi."
        
    def pay(self):
        return "you can pay by cash or electronic card."
    
    def total_fee(self):
        return self.__random_fee_per_km * self.random_distance_km
    
```
